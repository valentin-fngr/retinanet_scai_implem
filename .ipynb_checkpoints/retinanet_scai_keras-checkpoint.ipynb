- one-stage detectors that are applied over a regular, dense sampling of possible object locations have the potential to be faster and simpler, but have trailed the accuracy of two-stage detectors thus far. In this paper, we investigate why this is the case. We discover that the extreme foreground-background class imbalance encountered during training of dense detectors is the central cause

- We propose to address this class imbalance by reshaping the standard cross entropy loss such that it down-weights the loss assigned to well-classified examples => prevents the vast number of easy negatives from overwhelming the detector during training

# 1. RetinaNet Detector

- RetinaNet  is  a  single,  unified  network  composed  of  abackbone network and two task-specificsubnetworks

- the  first  subnet  performs  convo-lutional object classification on the backbone’s output; thesecond subnet performs convolutional bounding box regres-sion. 

- All new conv layers except the finalone in the RetinaNet subnets are initialized with biasb= 0and a Gaussian weight fill withσ= 0.01.

- or the final convlayer of the classification subnet, we set the bias initializa-tion  tob=−log((1−π)/π),  whereπspecifies  that  at the start of training every anchor should be labeled as fore-ground with confidence of∼π.  We useπ=.01

- For all ablation studies weuse an image scale of 600 pixels for training and testing

#### FPN : 

- Each level of the pyramidcan be used for detecting objects at a different scale.  FPNimproves multi-scale predictions from fully convolutionalnetworks (FCN) 

- As in [20] all pyramid levelshaveC= 256channels

- we emphasize the use of theFPN backbone is;  preliminary experiments using featuresfrom only the final ResNet layer yielded low AP

- We  construct  a  pyramid  with  levelsP3throughP7, wherelindicates pyramid level (Plhas reso-lution2llower than the input)

#### Anchors: 

- We use translation-invariant anchor boxes simi-lar to those in the RPN variant in [20].  The anchors haveareas of322to5122on pyramid levelsP3toP7,  respec-tively.  As in [20], at each pyramid level we use anchors atthree aspect ratios{1:2,1:1,2:1}.  For denser scale cover-age than in [20], at each level we add anchors of sizes{20,21/3,22/3}of the original set of 3 aspect ratio anchors

- Each  anchor  is  assigned  a  lengthKone-hot  vector  ofclassification  targets,  whereKis  the  number  of  objectclasses,  and a 4-vector of box regression targets.

- anchors are assigned to ground-truth object boxes using anintersection-over-union (IoU) threshold of 0.5; and to back-ground if their IoU is in [0, 0.4). 

- As each anchor is assignedto at most one object box, we set the corresponding entryin its lengthKlabel vector to1and all other entries to0.If an anchor is unassigned, which may happen with overlapin [0.4, 0.5), it is ignored during training

- Box regressiontargets are computed as the offset between each anchor andits assigned object box, or omitted if there is no assignment

#### Classification  Subnet: 

-  This subnetis a small FCN attached to each FPN level; parameters ofthis subnet are shared across all pyramid levels

- taking  an  input  feature  map  withCchannelsfrom  a  given  pyramid  level,  the  subnet  applies  four  3×3conv layers, each withCfilters and each followed by ReLUactivations, followed by a 3×3 conv layer withKAfilters

#### Box Regression Subnet

- The design of the box regression subnet is identical to theclassification subnet except that it terminates in4Alinearoutputs  per  spatial  location

-  For  eachof theAanchors per spatial location, these4outputs pre-dict the relative offset between the anchor and the ground-truth box (we use the standard box parameterization from R-CNN

## Anchor generation 

We will create a class that contains all important operation in order to retrieve anchors for each pyramid layer. We will detail all methods using a doc string

```python
class AnchorGenerator(): 
    
    def __init__(self, image_width, image_height): 
        self.ratios = [1/2, 1, 2] # the width is equal to the height multipled by ratio 
        self.scales = [2**0, 2**(1/3), 2**(2/3)]
        self.strides = list(range(3,8))
        self.area = [32**2, 64**2, 128**2, 256**2, 512**2]
        
        self.image_width = image_width 
        self.image_height = image_height
        self.dimensions = self.compute_dimensions()
        
    def compute_dimensions(self): 
        """
            retrieve all pyramid anchor dimensions by iterating over all pyramid levels and computing 
            their anchor dimensions. 
            Output: 
                (nb_pyramid_level, 9) <-> (5,9)
        """
        all_pyramid_anchors = []
        for area in self.area:
            
            pyramid_anchors = []
            
            for ratio in self.ratios: 
                h = tf.sqrt(area / ratio)
                w = area / h
                # scaling 
                for scale in self.scales: 
                    pyramid_anchors.append(tf.stack([w,h]) * scale)
            
            # retrieved 9 different anchor dimensions for a single pyramid level
            all_pyramid_anchors.append(pyramid_anchors)
                    
        return all_pyramid_anchors
    
    def retrieve_anchors_by_level(self, level): 
        """
            Given an anchor level, return all anchors, for all centers in the pyramid feature, 
            in a numpy array. 
            Argement: 
                level : a pyramid level 
            Ouptut: 
                a (feature_width * feature_height * 9, 4) tensor, containing all anchors for this pyramid ft 
            
        """
        
        dimension_index = level - 3
        feature_width, feature_height = (self.image_width / level**2),(self.image_height / level **2)
        feature_width = tf.cast(feature_width, dtype=tf.int32) 
        feature_height = tf.cast(feature_height, dtype=tf.int32)
        # create a grid over this feature pyramid 
        center_x, center_y = tf.meshgrid(tf.range(feature_width, dtype=tf.float32) + 0.5, tf.range(feature_height, dtype=tf.float32) + 0.5)
        centers = tf.stack([center_x, center_y], axis=-1) # stacking the center coordinates (fw, fh, 2)
        scaled_centers = centers * (2**self.strides[dimension_index]) # we are rescaling back !(fw, fh, 2)
        
        level_anchors = self.dimensions[dimension_index] # (9,2)
        level_anchors = np.reshape(level_anchors, [1, 1, 9, 2]) # (1,1,9,2)
        level_anchors = tf.tile(level_anchors, [feature_height, feature_width, 1, 1]) # (fw,fh,9,2)
        
        # in order to stack level_anchors w and h with the x,y of each center coords, 
        # we need to make sure that level_anchors and scaled_centers can be computer together
        # we add a dimension to scaled_centers 
        
        scaled_centers = tf.expand_dims(scaled_centers, axis=-2) # (fw, fh, 1, 2)
        scaled_centers = tf.tile(scaled_centers, [1,1,9,1]) # (fw, fh, 9, 2)
        # we now have shapes of scaled_centers : (fw, fh, 9, 2) and level_anchors : (fw,fh,9,2)
        
        scaled_centers = tf.cast(scaled_centers, dtype=tf.float32) 
        level_anchors = tf.cast(level_anchors, dtype=tf.float32)
        anchors_coords = tf.concat([scaled_centers, level_anchors], axis=-1) # (feature_width, feature_height, 9, 4)
        return tf.reshape(anchors_coords, [feature_width * feature_height * 9, 4])
        
        
    def retrieve_all_anchors(self): 
        
        anchors = [] 
        
        for level in self.strides: 
            anchors.append(self.retrieve_anchors_by_level(level))
            
        return tf.concat(anchors, axis=0)
```

```python
# testing our anchors generator ! 
import matplotlib.patches as patches

for sample in train_dataset.take(1): 
    image = sample["image"]
    anchor_gen = AnchorGenerator(*image.shape[:2])
    anchors = anchor_gen.retrieve_all_anchors()[:10] # printing the first 10 anchors
    
    fig, ax = plt.subplots(figsize=(10,10))
    
    for anchor in anchors: 
        rect = patches.Rectangle((anchor[0] - anchor[2] / 2 , anchor[1] - anchor[3] / 2), anchor[2], anchor[3], linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
        
    plt.imshow(image)
```
